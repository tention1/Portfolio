> We now want to look at the relationship between company and gross revenue; however, the company column features strings and not numeric values. We will first convert the company name to a numeric value...

> We could specify which column whose values we want to cast to a numeric value, but for simplicty we set the whole dataframe to a contain numeric values. This comes with the added benefit of allowing us to then analyze the other string-valued columns

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df.corr(method = 'pearson')

matplotlib.rcParams['figure.figsize'] = (12,8)

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation Matrix")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values(kind = 'quicksort')
print(sorted_pairs)
```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

> Most sources will tell us that a strong correlation applies to values for r that are > 0.7 or > 0.75. With this in mind, we see that only the budget and the gross revenue columns demonstrate strong enough correlation.
